# What are the main characteristics which have the most impact on a car price?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
%%capture
! pip install seaborn
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

```python
#understand what type of variable we are dealing with

print(df.dtypes)
```

```python
df_gp = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gp.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
grouped_test1.sort_values(by=['price'], inplace=True, ascending=False)
print(grouped_test1)
```

```python
# Descriptive Statistical Analysis
#The 'describe'function automatically computes basic statistics for all continuous variables

df.describe(include='all')
```

```python
#calculate the correlation between variables of type "int64" or "float64" using the method "corr"

df.corr()
```

```python
#Find the correlation between the following columns: bore, stroke,compression-ratio , and horsepower.

df[['bore', 'stroke', 'compression-ratio', 'horsepower']].corr()  
```

```python
# Engine size as potential predictor variable of price

sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0,)
df[['engine-size', 'price']].corr() 
```

```python
# Horsepower as potential predictor variable of price

sns.regplot(x="horsepower", y="price", data=df)
plt.ylim(0,)
df[['horsepower', 'price']].corr() 
```

```python
# Highway-mpg as potential predictor variable of price - negative linear correlation

sns.regplot(x="highway-mpg", y="price", data=df)
plt.ylim(0,)
df[['highway-mpg', 'price']].corr() 
```

```python
# Peak-rpm as potential predictor variable of price - weak linear correlation

sns.regplot(x="peak-rpm", y="price", data=df)
plt.ylim(0,)
df[['peak-rpm', 'price']].corr() 
```

```python
# Stroke as potential predictor variable of price - weak linear correlation

sns.regplot(x="stroke", y="price", data=df)
plt.ylim(0,)
df[['stroke', 'price']].corr() 
```

```python
# Categorical variables correlation with Boxplots

sns.boxplot(x="body-style", y="price", data=df)
```
